```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("C:/Users/SOUVIK/Downloads/social_media_sentiment.csv")
```

```python
# Print dataset overview
print(df.head())
```

```python
# Count of sentiments
sentiment_counts = df['sentiment'].value_counts()
```

```python
# Bar plot of sentiment distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='sentiment', palette='Set2', order=sentiment_counts.index)
plt.title("Sentiment Distribution in Social Media Posts")
plt.xlabel("Sentiment")
plt.ylabel("Number of Posts")
plt.tight_layout()
plt.show()
```
