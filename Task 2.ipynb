```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("C:/Users/SOUVIK/Downloads/titanic_sample.csv")
```

```python
# Show basic info
print(df.info())
print(df.describe(include='all'))
```

```python
# Check for missing values
print("Missing Values:\n", df.isnull().sum())
```

```python
# Convert 'Survived' to category
df['Survived'] = df['Survived'].map({0: 'No', 1: 'Yes'})
```

```python
# 1. Survival count
sns.countplot(data=df, x='Survived', palette='Set2')
plt.title('Survival Count')
plt.show()
```

```python
# 2. Survival by Gender
sns.countplot(data=df, x='Sex', hue='Survived', palette='Set1')
plt.title('Survival by Gender')
plt.show()
```

```python
# 3. Age distribution
sns.histplot(data=df, x='Age', bins=5, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# 4. Fare distribution by class
sns.boxplot(data=df, x='Pclass', y='Fare', palette='Pastel1')
plt.title('Fare Distribution by Class')
plt.show()
```
