```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("C:/Users/SOUVIK/Downloads/us_traffic_accidents_sample.csv")
```

```python
# Convert Start_Time to datetime
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
```

```python
# Extract hour of accident
df['Hour'] = df['Start_Time'].dt.hour
```

```python
# Plot 1: Severity by Weather Condition
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Weather_Condition', y='Severity', palette='coolwarm')
plt.title('Severity vs Weather Condition')
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Accidents by Time of Day
plt.figure(figsize=(8,5))
sns.histplot(df['Hour'], bins=8, kde=True)
plt.title('Distribution of Accidents by Hour')
plt.xlabel('Hour of Day')
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Visibility vs Severity
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Visibility(mi)', y='Severity', hue='Weather_Condition')
plt.title('Visibility vs Severity by Weather Condition')
plt.tight_layout()
plt.show()
```
